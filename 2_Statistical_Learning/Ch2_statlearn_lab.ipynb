## Indexing Data
To begin, we  create a two-dimensional `numpy` array.

````python
A = np.array(np.arange(16)).reshape((4, 4))
A
````

Typing `A[1,2]` retrieves the element corresponding to the second row and third
column. (As usual, `Python` indexes from $0.$)

````python
A[1,2]
````

The first number after the open-bracket symbol `[`
 refers to the row, and the second number refers to the column.

### Indexing Rows, Columns, and Submatrices
 To select multiple rows at a time, we can pass in a list
  specifying our selection. For instance, `[1,3]` will retrieve the second and fourth rows:

````python
A[[1,3]]
````

To select the first and third columns, we pass in  `[0,2]` as the second argument in the square brackets.
In this case we need to supply the first argument `:`
which selects all rows.

````python
A[:,[0,2]]
````

Now, suppose that we want to select the submatrix made up of the second and fourth
rows as well as the first and third columns. This is where
indexing gets slightly tricky. It is natural to try  to use lists to retrieve the rows and columns:

````python
A[[1,3],[0,2]]
````

 Oops --- what happened? We got a one-dimensional array of length 2 identical to

````python
np.array([A[1,0],A[3,2]])
````

 Similarly,  the following code fails to extract the submatrix comprised of the second and fourth rows and the first, third, and fourth columns:

````python
# A[[1,3],[0,2,3]]

# ---------------------------------------------------------------------------
# IndexError                                Traceback (most recent call last)
# Cell In[64], line 1
# ----> 1 A[[1,3],[0,2,3]]

# IndexError: shape mismatch: indexing arrays could not be broadcast together with shapes (2,) (3,) 
````

We can see what has gone wrong here. When supplied with two indexing lists, the `numpy` interpretation is that these provide pairs of $i,j$ indices for a series of entries. That is why the pair of lists must have the same length. However, that was not our intent, since we are looking for a submatrix.

One easy way to do this is as follows. We first create a submatrix by subsetting the rows of `A`, and then on the fly we make a further submatrix by subsetting its columns.


````python
A[[1,3]][:,[0,2]]
````

There are more efficient ways of achieving the same result.

The *convenience function* `np.ix_()` allows us  to extract a submatrix
using lists, by creating an intermediate *mesh* object.

````python
idx = np.ix_([1,3],[0,2,3])
A[idx]
````

Alternatively, we can subset matrices efficiently using slices.
  
The slice
`1:4:2` captures the second and fourth items of a sequence, while the slice `0:3:2` captures
the first and third items (the third element in a slice sequence is the step size).

````python
A[1:4:2,0:3:2]
````

Why are we able to retrieve a submatrix directly using slices but not using lists?
Its because they are different `Python` types, and
are treated differently by `numpy`.
Slices can be used to extract objects from arbitrary sequences, such as strings, lists, and tuples, while the use of lists for indexing is more limited.




    



    



### Boolean Indexing
In `numpy`, a *Boolean* is a type  that equals either   `True` or  `False` (also represented as $1$ and $0$, respectively).
The next line creates a vector of $0$'s, represented as Booleans, of length equal to the first dimension of `A`.

````python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
````

We now set two of the elements to `True`.

````python
keep_rows[[1,3]] = True
keep_rows
````

Note that the elements of `keep_rows`, when viewed as integers, are the same as the
values of `np.array([0,1,0,1])`. Below, we use  `==` to verify their equality. When
applied to two arrays, the `==`   operation is applied elementwise.

````python
np.all(keep_rows == np.array([0,1,0,1]))
````

(Here, the function `np.all()` has checked whether
all entries of an array are `True`. A similar function, `np.any()`, can be used to check whether any entries of an array are `True`.)

   However, even though `np.array([0,1,0,1])`  and `keep_rows` are equal according to `==`, they index different sets of rows!
The former retrieves the first, second, first, and second rows of `A`.

````python
A[np.array([0,1,0,1])]
````

 By contrast, `keep_rows` retrieves only the second and fourth rows  of `A` --- i.e. the rows for which the Boolean equals `TRUE`.

````python
A[keep_rows]
````

This example shows that Booleans and integers are treated differently by `numpy`.

## For Loops
A `for` loop is a standard tool in many languages that
repeatedly evaluates some chunk of code while
varying different values inside the code.
For example, suppose we loop over elements of a list and compute their sum.

````python
total = 0
for value in [3,2,19]:
    total += value
print('Total is: {0}'.format(total))
````

The indented code beneath the line with the `for` statement is run
for each value in the sequence
specified in the `for` statement. The loop ends either
when the cell ends or when code is indented at the same level
as the original `for` statement.
We see that the final line above which prints the total is executed
only once after the for loop has terminated. Loops
can be nested by additional indentation.

````python
total = 0
for value in [2,3,19]:
    for weight in [3, 2, 1]:
        total += value * weight
print('Total is: {0}'.format(total))
````

Above, we summed over each combination of `value` and `weight`.
We also took advantage of the *increment* notation
in `Python`: the expression `a += b` is equivalent
to `a = a + b`. Besides
being a convenient notation, this can save time in computationally
heavy tasks in which the intermediate value of `a+b` need not
be explicitly created.

Perhaps a more
common task would be to sum over `(value, weight)` pairs. For instance,
to compute the average value of a random variable that takes on
possible values 2, 3 or 19 with probability 0.2, 0.3, 0.5 respectively
we would compute the weighted sum. Tasks such as this
can often be accomplished using the `zip()`  function that
loops over a sequence of tuples.

````python
total = 0
for value, weight in zip([2,3,19],
                         [0.2,0.3,0.5]):
    total += weight * value
print('Weighted average is: {0}'.format(total))
````

**(a) Sketch**

* **X-axis:** Model flexibility (left = inflexible, right = highly flexible).
* **Y-axis:** Error or bias/variance magnitude.

**Curves:**

1. **Bias²:** Starts high (inflexible models underfit), decreases steadily, approaches zero with more flexibility.
2. **Variance:** Starts low (stable estimates for simple models), increases sharply with flexibility (overfitting).
3. **Training error:** Decreases monotonically (more flexibility = better fit to training data).
4. **Test error:** U-shaped — high for inflexible (underfit), lowest at some middle flexibility, then rises again (overfit).
5. **Bayes error (irreducible error):** Flat horizontal line above zero — represents noise in the data that no method can reduce.


````python
import numpy as np
import matplotlib.pyplot as plt

# Flexibility axis
x = np.linspace(0, 10, 200)

# Curves (stylized)
bias2 = (10 - x)**2 / 25        # decreasing curve
variance = x**2 / 25            # increasing curve
train_error = 5 / (x + 1)       # monotonically decreasing
test_error = bias2 + variance + 2  # U-shaped (bias^2 + variance + noise)
bayes_error = np.ones_like(x) * 2  # flat irreducible error

plt.figure(figsize=(8,6))

plt.plot(x, bias2, label="Bias²", linewidth=2)
plt.plot(x, variance, label="Variance", linewidth=2)
plt.plot(x, train_error, label="Training Error", linewidth=2)
plt.plot(x, test_error, label="Test Error", linewidth=2)
plt.plot(x, bayes_error, label="Bayes Error", linewidth=2, linestyle="--")

plt.xlabel("Model Flexibility →", fontsize=12)
plt.ylabel("Error / Variance / Bias²", fontsize=12)
plt.title("Bias-Variance Tradeoff and Error Curves", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````

* **Training error:** Always decreases as flexibility grows (a flexible model can fit training points more closely, possibly perfectly).
* **Test error:** Combines bias² + variance + irreducible error. At low flexibility, bias dominates → high test error. At high flexibility, variance dominates → also high error. Minimum occurs in between (sweet spot).
* **Bayes error:** Irreducible — reflects inherent randomness/noise in $Y$. Cannot be reduced by any model, hence a flat line.

5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?


**Advantages of a very flexible approach**

* **Captures complex relationships:** Can model highly nonlinear or intricate patterns in the data that simpler methods would miss.
* **Lower bias:** Flexible models make fewer assumptions about the functional form, so they can better approximate the true underlying relationship.
* **Potential for higher predictive accuracy:** Especially when the true data-generating process is complex and the sample size is large.

**Disadvantages of a very flexible approach**

* **High variance / overfitting:** Flexible methods can adapt too closely to the training data, capturing noise as if it were signal.
* **Interpretability issues:** Models like deep neural networks or highly tuned ensembles are often “black boxes,” making inference about variable effects difficult.
* **Data hungry:** Require large sample sizes to estimate parameters reliably without overfitting.
* **Computational cost:** Often more demanding in terms of training time and resources.

**When a more flexible approach is preferred**

* When the true relationship between predictors and response is highly nonlinear or complex.
* When the number of observations $n$ is large relative to the number of predictors $p$, reducing variance risk.
* When *prediction accuracy* is the main objective, rather than interpretability.

**When a less flexible approach is preferred**

* When the sample size is small or $p$ is large relative to $n$.
* When the true relationship is simple or approximately linear.
* When the goal is *inference* (understanding which predictors matter and how they affect the response).
* When interpretability, stability, and generalizability are valued over squeezing out small gains in predictive performance.

6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a non-parametric approach)? What are its disadvantages?


**Parametric vs. Non-Parametric Approaches**

* **Parametric methods** (e.g., linear regression, logistic regression, LDA):

  * Assume a specific functional form for the relationship between predictors and response (e.g., linear, logistic).
  * The task is to estimate a finite set of parameters within this form.
  * Once parameters are estimated, predictions are straightforward and efficient.

* **Non-parametric methods** (e.g., k-nearest neighbors, decision trees, splines):

  * Do **not** assume a fixed functional form.
  * Instead, they let the data “speak for itself,” fitting more flexible relationships.
  * Can approximate very complex functions as sample size grows.


**Advantages of Parametric Approaches**

1. **Simplicity and interpretability:** Easy to understand and explain results (e.g., regression coefficients).
2. **Data efficiency:** Work well with small sample sizes, since only a few parameters need to be estimated.
3. **Lower variance:** Because the model is constrained, it is less sensitive to noise.
4. **Faster computation:** Estimation and prediction are typically more computationally efficient.

**Disadvantages of Parametric Approaches**

1. **High bias if assumptions are wrong:** If the true relationship is nonlinear or complex, parametric methods will underfit.
2. **Rigid structure:** Lack flexibility to adapt to unusual patterns in the data.
3. **Model misspecification risk:** If the chosen functional form is inappropriate, predictions and inferences will be poor.


**Advantages of Non-Parametric Approaches**

* Greater flexibility: can capture nonlinear and complex relationships.
* Fewer assumptions: less risk of model misspecification.

**Disadvantages of Non-Parametric Approaches**

* Require much larger sample sizes to avoid overfitting.
* Typically less interpretable.
* Can be computationally intensive.

   (d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the *best* value for $K$ to be large or small? Why?

 (a) Euclidean distances from the test point $(0,0,0)$

Distance formula: $d = \sqrt{(X_1-0)^2 + (X_2-0)^2 + (X_3-0)^2} = \sqrt{X_1^2 + X_2^2 + X_3^2}$.

Compute for each observation:

1. Obs 1: $(0,3,0)$ → $d = \sqrt{0^2 + 3^2 + 0^2} = \sqrt{9} = 3.0000$.
2. Obs 2: $(2,0,0)$ → $d = \sqrt{2^2 + 0^2 + 0^2} = \sqrt{4} = 2.0000$.
3. Obs 3: $(0,1,3)$ → $d = \sqrt{0^2 + 1^2 + 3^2} = \sqrt{10} \approx 3.1623$.
4. Obs 4: $(0,1,2)$ → $d = \sqrt{0^2 + 1^2 + 2^2} = \sqrt{5} \approx 2.2361$.
5. Obs 5: $(-1,0,1)$ → $d = \sqrt{(-1)^2 + 0^2 + 1^2} = \sqrt{2} \approx 1.4142$.
6. Obs 6: $(1,1,1)$ → $d = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3} \approx 1.7321$.

Sorted by distance (closest first):
Obs 5 (≈1.4142, **Green**), Obs 6 (≈1.7321, **Red**), Obs 2 (2.0000, **Red**), Obs 4 (≈2.2361, **Green**), Obs 1 (3.0000, **Red**), Obs 3 (≈3.1623, **Red**).

 (b) Prediction with $K=1$

The single nearest neighbor is **Obs 5**, whose class is **Green**.
**Prediction:** **Green.**
**Why:** K=1 uses the class of the closest training observation.

 (c) Prediction with $K=3$

The 3 nearest neighbors are Obs 5 (Green), Obs 6 (Red), and Obs 2 (Red). That gives **2 Reds** and **1 Green**.
**Prediction:** **Red.**
**Why:** K=3 predicts by majority vote among the 3 nearest neighbors.

(d) If the Bayes decision boundary is highly non-linear, should the best $K$ be large or small? Why?

If the true Bayes decision boundary is highly non-linear, we want a classifier that can produce a flexible, wiggly boundary to approximate those complexities. **A small $K$** yields a more flexible, locally adaptive decision boundary (less smoothing), so it can capture intricate non-linear structure. A **large $K$** smooths the classification heavily (less flexible) and is likely to miss fine-grained non-linear patterns.
**Therefore:** the best $K$ would tend to be **small** when the Bayes boundary is highly non-linear.
